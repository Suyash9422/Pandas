Question : Find the 3 teams against which virat kolhi scored max runs.

```python
mask1=ipl['batsman']=='V Kohli'
kolhi=ipl[mask1]
```

```python
kolhi
```

```python
kolhi.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def max_runs(batsman_name):
    mask1=ipl['batsman']==batsman_name
    kolhi=ipl[mask1]
    return kolhi.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
max_runs('AB de Villiers')
```

```python
#find the most destrcutive death over batsman in the history of ipl
#strike rate => (number of runs/number of balls)*100
#min batsman 200 balls in over 16-20 over
```

```python
ipl=pd.read_csv('deliveries.csv')
```

```python
death_over=ipl[ipl['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
# now we need to find number of runs and the number of balls of the shortlisted players (i.e 43 batsman)
batsman_list
```

```python
final=ipl[ipl['batsman'].isin(batsman_list)]
final
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```
