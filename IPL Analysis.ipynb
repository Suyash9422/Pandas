This method will sort the data into ascending or descending order

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False)
```

```python
data.sort_values('season')
```

# Drop_duplicates () Method

This method will remove duplicate records or values from the data. We will keep duplicates as it is then the efficiency of the model will decrease , so it is imp to remove duplicate values or records.

```python
data
```

```python
data.drop_duplicates(subset=['city'])
```

```python
data.drop_duplicates(subset=['city']).shape
```

```python
data.drop_duplicates(subset=['city'],keep='first')
```

```python
data.drop_duplicates(subset=['city'],keep='last')
```

```python
data.drop_duplicates(subset=['city','season']).shape
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']]
```
